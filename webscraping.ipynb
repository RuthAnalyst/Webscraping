```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.homely.com.au/sold-properties/milang-sa-5256"
page  = requests.get(url)

soup = BeautifulSoup(page.text, "html.parser")

soup1 = soup.prettify()

print(soup1)
```

```python
tag = soup.findAll("h2", {"class": "sc-9e4dd548-1 hUXfPl"})


print(tag)
```

```python
address = []
for each in tag: address.append((each.text.strip()))
print(address)
```

```python
import csv
```

```python
with open('address', 'w') as f:
    write = csv.writer(f)
    write.writerows(address)
```

```python
df = pd.DataFrame(address)
print(df)
```

```python
df.to_csv("C:/Users/Mum/Downloads/address.csv", index=False)
```
